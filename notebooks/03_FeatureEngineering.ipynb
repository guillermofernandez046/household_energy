# feature enginerring

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler #normalizacion de caracteristicas
```

```python
base_fe = pd.read_csv('C:/Users/MyDELL/Desktop/msds/repositorio/household_energy/datasets/base_dw.csv')
```

```python
base_fe.head()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Temperature',
    y='EnergyConsumption',
    data=base_fe,
    hue='HVACUsage',
    palette='Set1',    
    alpha=0.7
)

plt.title('Consumo de Energa vs Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Consumo de Energa (kWh)')
plt.legend(title='HVAC Usage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# normalizacion de valores, se normalizan los valores para que el modelo pueda procesar ams rapido

```python
escalar_cols = ['Temperature', 'SquareFootage', 'Occupancy', 'RenewableEnergy']
scaler = StandardScaler()
base_fe[escalar_cols] = scaler.fit_transform(base_fe[escalar_cols])
```

```python
base_fe.head()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Temperature',
    y='EnergyConsumption',
    data=base_fe,
    hue='HVACUsage',
    palette='Set1',    
    alpha=0.7
)

plt.title('Consumo de Energa vs Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Consumo de Energa (kWh)')
plt.legend(title='HVAC Usage')
plt.grid(True)
plt.tight_layout()
plt.show()
```
